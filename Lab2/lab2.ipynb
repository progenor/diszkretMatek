```python
def readnumbersVertical2():
    with open("valami2.txt", 'r') as f:
        lines = f.read()
    lines = lines.split('\n')
    numbers = []
    for i in lines:
        try:
            for i in i.split():
                numbers.append(int(i))
        except:
            numbers.append(int(i)) 
    return numbers
readnumbersVertical2()
```

```python
def readnumbersVertical():
    with open("valami2.txt", 'r') as f:
        lines = f.read()
    lines = lines.split('\n')
    numbers = [int(i) for i in lines]
    return numbers
readnumbersVertical()
```

```python
def readInputWithOpen():
    with open("valami.txt", "r") as f:
        line = f.readline().strip()
    numbers = [int(i) for i in line.split()]
    return numbers  
readInputWithOpen()
```
